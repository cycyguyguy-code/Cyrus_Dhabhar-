```python
sales = float(input("Enter the Sales ($): "))
taxrate = float(input("Enter the Tax Rate: *(%) "))

cogs = 0.4 * sales
grossprofit = sales - cogs
tax = (taxrate/100)* grossprofit
netincome = grossprofit - tax

print(f"sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"grossprofit {grossprofit}$")
print(f"tax: {tax}$")
print(f"Net income: {netincome}$")

```
