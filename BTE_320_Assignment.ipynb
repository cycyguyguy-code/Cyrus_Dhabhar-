```python
def roi(cost, revenue):
  assert isinstance(cost, (int, float)), "Error: Cost must be a numeric value."
  assert isinstance(revenue, (int, float)), "Error: Revenue must be a numeric value."
  assert cost > 0, "Error: Cost must be greater than 0."
  return (revenue - cost)/cost
```

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Error: Principal must be a numeric value."
  assert isinstance(annual_rate, (int, float)), "Error: Annual rate must be a numeric value."
  assert isinstance(months, int), "Error: Months must be an integer."
  if (principal <= 0) or (annual_rate <= 0) or (months <= 0):
    raise ValueError("Error: Principal, annual rate, and months must be positive numbers.")

  try:
    r = annual_rate / 12
    if r == 0:
      return principal / months
    else:
      return principal * r / ((1 - (1 + r))** (-months))
  except Exception as e:
    print(f"Error: {e}")
```

```python
def getValues(n):
  assert isinstance(n, int), "Input variable n must be an integer."
  assert n > 0
  items = []
  for i in range(n):
    while True:
      try:
        value = float(input(f"Add value {i+1}/{n} to the list: "))
        items.append(value)
        break
      except ValueError:
        print("Invalid input. Please enter a valid number.")
  return items

def calcAverage(alist):
  assert isinstance(alist, list)
  assert len(alist) > 2
  try:
    return sum(alist[2:]) / len(alist[2:])
  except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.")
  except:
    raise ValueError("The length of list 'alist' must be greater than 2")

def main():
  n = int(input("Enter the number of items: "))
  alist = getValues(n)
  avg = calcAverage(alist)
  print(avg)

main()
```

```python
class Student:
    def __init__(self, name, number):
        self.__name = name
        self.number = number
        self.__courses = []

    def get_name(self):
        return self.__name

    def get_courses(self):
        return self.__courses

    def set_name(self, newName):
        self.__name = newName

    def __len__(self):
        return len(self.__courses)

    def enroll(self, newCourse):
        if newCourse not in self.__courses:
            self.__courses.append(newCourse)
            print(f"Enrolled in {newCourse}")
        else:
            print(f"Student {self.__name} has already enrolled in {newCourse}")


s = Student("Jane Doe", 12345)
print(f"Student name: {s.get_name()}")
print(f"Show currently enrolled courses: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Show currently enrolled courses: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Total number of courses enrolled: {len(s)}")
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5

class Rectangle:
    def __init__(self, p1, p2, p3, p4):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4
        self.width = self.width_()
        self.height = self.height_()

    def width_(self):
        return self.p1.distance(self.p2)

    def height_(self):
        return self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height

p1 = Point(0, 0)
p2 = Point(5, 0)
p3 = Point(0, 5)
p4 = Point(5, 5)

z = Rectangle(p1, p2, p3, p4)
z.area()
```

```python
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self.__hourly_wage = wage
        self.hours = hours

    def get_wage(self):
        return self.__hourly_wage

    def set_wage(self, newWage):
        assert newWage > 0, "wage values cannot be negative or zero."
        self.__hourly_wage = newWage

    def calculate_pay(self):
        return self.__hourly_wage * self.hours

    def __str__(self):
        return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}$/hour"

class Manager(Employee):
    def __init__(self, name, wage, hours, bonus):
        super().__init__(name, wage, hours)
        self.bonus = bonus # assume bonus given as float 0 - 1 (20% is 0.2)

    def calculate(self):
        return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Allice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee pay: {employee.name} weekly salary is {employee.calculate_pay()}$")

print(f"Manager pay: {manager.name} weekly salary is {manager.calculate()}$ that includes a {manager.bonus * 100}% bonus")
```
