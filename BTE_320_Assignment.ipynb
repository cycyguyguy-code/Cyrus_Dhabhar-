```python
sales = float(input("Enter the Sales ($): "))
taxrate = float(input("Enter the Tax Rate: *(%) "))

cogs = 0.4 * sales
grossprofit = sales - cogs
tax = (taxrate/100)* grossprofit
netincome = grossprofit - tax

print(f"sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"grossprofit {grossprofit}$")
print(f"tax: {tax}$")
print(f"Net income: {netincome}$")

```

```python
num_1 = float(input("Enter the first number: "))
operator = input("Enter Numerica operator (+, -, *, /, **, %): ")
num_2 = float(input("Enter the second number: "))

result = None

if operator == '+':
    result = num_1 + num_2
elif operator == '-':
    result = num_1 - num_2
elif operator == '*':
    result = num_1 * num_2
elif operator == '**':
    result = num_1 ** num_2
elif operator == '%':

  result = num_1 % num_2
elif operator == '/':
    if num_2 != 0:
        result = num_1 / num_2
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Error: Invalid operator.")


print(f"The result is: {result}")
```

```python
print("Hour" , "       Fee")
for hours in range(1,9):
  fee = 5 + (2.5*(hours))
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(hours, "hours:", "$",fee)
```

```python
ticketPrice = 10
inititAttendees = 20

print( "Ad$ Profit")
for ad in range (0, 201, 25 ):
  revenue = (inititAttendees + 2 * round(ad ** 0.5) )* ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)} {float(profit)}")
```

```python
ticketPrice = 10
inititAttendees = 20

print( "Ad$ Profit")
ad = 0
while ad < 201 :
  revenue = (inititAttendees + 2 * round(ad ** 0.5) )* ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)} {float(profit)}")
  ad += 25
```

```python
import random

check = input("Do you want to play a game (y/n)?")

while check == 'y' :
  player = input("Choose between rock, paper, scissors:").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player}")
  print(f"Computer: {computer}")

  if player == computer:
    print('Draw')
  elif ( (player == 'paper' and computer == 'rock') or
      (player == 'scissors' and computer == 'paper') or
      (player == 'rock' and computer == 'scissors') ):
    print("Player wins")
  else:
    print("computer wins")

  check = input("Do you want to play again (y/n)?").lower()
```
